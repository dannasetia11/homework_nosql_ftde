```python
import pymongo
from datetime import date, datetime
import math
import random
from bson.objectid import ObjectId

myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["dskola"]
mycol = mydb["books"]

def generate_valid_date(year):
    while True:
        try:
            month = random.randint(1, 12)
            day = random.randint(1, 28) if month == 2 else random.randint(1, 30) if month in [4, 6, 9, 11] else random.randint(1, 31)
            return datetime(year, month, day)
        except ValueError:
            continue


for i in range(0, 100) :

    my_dict = {
        "title": "Book Title " + str(i),
        "author": "Author " + str((i % 10 + 1)), # 10 different authors
        "genre": random.choice(["Fiction", "Dystopian", "Adventure", "Science Fiction", "Fantasy"]), # 5 different genres
        "published_year": 1950 + (i % 70), # Published between 1950 and 2019
        "copies_available": math.floor(random.randint(1, 9) * 10) + 1, # 1 to 10 copies available
        "borrowed_by": [
            {
                "user_id": ObjectId(),
                "borrowed_date": generate_valid_date(2024), # Borrowed dates spread over the year 2024
                "return_date": None
            }
        ]
    }
    mycol.insert_one(my_dict)
```

```python
import pymongo

# Koneksi ke MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["dskola"]
collection = db["books"]

# Membaca semua dokumen dari koleksi
documents = collection.find()

# Menampilkan hasil
for doc in documents:
    print(doc)
```

```python
filterNo1 = collection.find({ "author": "Author 5",
    "borrowed_by": {
        "$elemMatch": {
            "return_date": None
        }
    }} )

for doc in filterNo1:
    print(doc)
```

```python
filterNo2 = collection.find({"published_year": {"$lt": 1980},
    "copies_available": {"$gt": 6}
                             })

for doc in filterNo2:
    print(doc)
```

```python
filterNo3 = collection.aggregate({"genre": "Fantasy"}).limit(5)

for doc in filterNo3:
    print(doc)
```

```python
filterNo4 = collection.aggregate([
    {
        "$group": {
            "_id": "$genre",
            "count": {"$sum": 1}
        }
    },
    {
        "$sort": {"count": -1}
    }
])

for result in filterNo4:
    print(f"Genre: {result['_id']}, Count: {result['count']}")
```

```python
filterNo5 = collection.find({
    "borrowed_by": []
})

for doc in filterNo5:
    print(doc)
```
